# Q1.  Find Numbers with Even Number of Digits
Given a valid (IPv4) IP address, return a defanged version of that IP address.
A defanged IP address replaces every period "." with "[.]".

```python
class Solution(object):
    def defangIPaddr(self, address):
        """
        :type address: str
        :rtype: str
        """
        self.address=address
        a=self.address.replace(".","[.]")
        return a

ip="1.2.5.4.3"
obj1=Solution()
print(obj1.defangIPaddr(ip))
```

# Q2. Find Numbers with Even Number of Digits
    Given an array nums of integers, return how many of them contain an even number of digits.

```python

class Solution(object):
    def findNumbers(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        even=0
        for i in nums:
            def count(val):
                if val//10==0:
                    return 1
                return 1+count(val//10)
            c=count(i)
            if c%2==0:
                even+=1
        return even        
        
list1= [233,13,4,33,222]
obj= Solution()
print(obj.findNumbers(list1))
                
```

# Q3. Number of Good Pairs
    Given an array of integers nums, return the number of good pairs.
    A pair (i, j) is called good if nums[i] == nums[j] and i < j.

```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # self.nums=nums
        count=0
        val=len(nums)
        for i in range(val):
            for j in range(val):
                if nums[i]==nums[j] and i<j:
                    count+=1
        return count
lst1=[1,2,4,2,4,2,2]
obj=Solution()
obj.numIdenticalPairs(lst1)
```

# Q4. How Many Numbers Are Smaller Than the Current Number 
    Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. That is, for each nums[i] you have to count the      number of valid j's such that j != i and nums[j] < nums[i].
    Return the answer in an array.

```python
class Solution(object):
    def smaller(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        list1=[]
        l=len(nums)
        for i in range(l):
            count=0
            for j in range(l):
                if nums[i]>nums[j]:
                    count+=1
            list1.append(count)
        return list1
list1=[33,33,22,432,55,32,66,77,34]
obj1=Solution()
print(obj1.smaller(list1))
```

# Q5. Subtract the Product and Sum of Digits of an Integer
    Given an integer number n, return the difference between the product of its digits and the sum of its digits.

```python
class Solution(object):
    def subtractProductAndSum(self, n):
        """
        :type n: int
        :rtype: int
        """
        sum = 0
        mul = 1
        while n>0:
            sum += n%10 
            mul *= n%10
            n //= 10       
        return mul-sum
obj=Solution()
print(obj.subtractProductAndSum(2644))
```

# Q6. XOR Operation in an Array
    You are given an integer n and an integer start.
    Define an array nums where nums[i] = start + 2 * i (0-indexed) and n == nums.length.
    Return the bitwise XOR of all elements of nums.

```python
class Solution(object):
    def xorOperation(self, n, start):
        """
        :type n: int
        :type start: int
        :rtype: int
        """
        result=0
        for i in range(n):
            val=start+2*i
            result^=val
        return result
obj=Solution()
print(obj.xorOperation(4,3))
```
